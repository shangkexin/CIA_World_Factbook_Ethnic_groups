```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select 

import pandas as pd
import numpy as np
import sqlite3
import re
```

# Web scraping

```python
driver = webdriver.Chrome(r'C:\Users\panda\Downloads\chromedriver-win32\chromedriver')
# get to the website
driver.get(f"https://www.cia.gov/the-world-factbook/field/ethnic-groups/")

texts = driver.find_elements(By.CLASS_NAME, 'pb30')
```

## Ethnic1

```python
Location = []
val = []
for t in texts:
    str = t.text
    try:
        country = str.split('\n')[0]
        Location.append(country)
    except:
        Location.append(np.nan)

    numeric_const_pattern = r"""
        [-+]? # optional sign
        (?:
            (?: \d* \. \d+ ) # .1 .12 .123 etc 9.1 etc 98.1 etc
            |
            (?: \d+ \.? ) # 1. 12. 123. etc 1 12 123 etc
        )
        # followed by optional exponent part if desired
        (?: [Ee] [+-]? \d+ ) ?
        """
    rx = re.compile(numeric_const_pattern, re.VERBOSE)
    x = rx.findall(str)
    try:
        value = x[0]
        # print(value)
        val.append(value)
    except:
        val.append(np.nan)


    
    # try:
    #     x = re.findall("\d+\.\d+", str) #find all the floating numbers in the string
    #     value = float(x[0])
    #     print(value)
    #     if value <= 100:
    #         val.append(value)
    #     else:
    #         val.append(np.nan)
    # except:
    #     val.append(np.nan)
    # else:
    #     x = re.findall('[0-9]+', str) #find all the integers in the string
    #     value = int(x[0])
    #     print(value)
    #     if value <= 100:
    #         val.append(value)
    #     else:
    #         val.append(np.nan)

df = pd.DataFrame({'Location': Location,
                  'val': val})
df.val = df.val.apply(pd.to_numeric, errors='coerce')
df = df[~df.Location.isna()]
df = df[~df.val.isna()]
df = df[df['val'] <= 100]
df.rename(columns={'val': '2024'}, inplace=True)
df.loc[df['Location'] == 'Taiwan', '2024'] = 70
df.loc[df['Location'] == 'United Arab Emirates', '2024'] = 59.4
df.loc[df['Location'] == 'Suriname', '2024'] = 27.4
df.loc[df['Location'] == 'Malaysia', '2024'] = 52.8
df.loc[df['Location'] == 'Kuwait', '2024'] = 40.3

country = pd.read_excel('Country.xlsx')
df = country.merge(df,
                   on = 'Location',
                   how = 'left')
df = df.drop(axis=1, columns=['Location'])
df

# Then merge with the old data in IFsHistSeries
```
